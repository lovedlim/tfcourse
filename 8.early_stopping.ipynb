```python
# 라이브러리 불러오기
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Mnist 데이터셋
mnist = tf.keras.datasets.mnist
(X, y), (x_test, y_test) = mnist.load_data()

# 시퀀셜 모델 만들기
model = Sequential([
                    Flatten(input_shape=(28,28)),
                    Dense(64, activation='relu'),
                    Dense(10, activation='softmax')
])

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 조기종료와 콜백 
# 콜백은 학습 실행 시 손실값 등을 모니터링 하고 손실에 따라 일부 작업을 수행할 수 있음

from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_accuracy', patience=3) #monitor: val_accuracy 관찰, patience: 에폭 3번까지 개선되지 않을 시 종료
# early_stopping = EarlyStopping(monitor='val_loss', patience=3) #val_loss 관찰
history = model.fit(X, y, validation_split=0.2, epochs=30, batch_size=64, callbacks=[early_stopping])
```

```python
# 새모델 만들기
model = Sequential([
                    Flatten(input_shape=(28,28)),
                    Dense(64, activation='relu'),
                    Dense(10, activation='softmax')
])

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# min_delta 기준 미달시 조기 종료
# 3epoch 동안 손실이 0.05이상 개선되지 않을 경우 훈련 중지
early_stopping = EarlyStopping(monitor='val_loss', patience=3, min_delta=0.05)
history = model.fit(X, y, validation_split=0.2, epochs=30, batch_size=64, callbacks=[early_stopping])
```
